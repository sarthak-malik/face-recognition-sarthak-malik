```
import cv2
def frame(path):
  cap= cv2.VideoCapture(path)
  i=0
  while(cap.isOpened()):
      ret, frame = cap.read()
      if ret == False:
          break
      cv2.imwrite("/content/frames/"+str(i)+".jpg",frame)
      
      i+=1
  
  cap.release()
  
```

```
#extracting
from PIL import Image as IM

def cord_face(imagePath,j):
    image = cv2.imread(imagePath)
    gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

    faceCascade = cv2.CascadeClassifier(cv2.data.haarcascades + "haarcascade_frontalface_default.xml")
    faces = faceCascade.detectMultiScale(
        gray,
        scaleFactor=1.3,
        minNeighbors=3,
        minSize=(30, 30)
    )
    extract_face(faces,imagePath,j)
  
def extract_face(faces,imagePath,j):
  
  image = cv2.imread(imagePath)
  for (x, y, w, h) in faces:
    roi_color = image[y:y + h, x:x + w] 
    
    gray = cv2.cvtColor(roi_color, cv2.COLOR_BGR2GRAY)
    I=IM.fromarray(gray)
    I.save("/content/data/me/"+str(j)+".jpg")
    j=j+1
```

```
lst=os.listdir("/content/frames/")
j=0
for img in lst:
  cord_face("/content/frames/"+img,j)
  j=j+1
```
